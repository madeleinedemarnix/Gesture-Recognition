```
class_id = {}
user_id = {}

files = [1, 11, 21, 101, 111, 121, 201, 211, 221]
for file in files:
    with open('/work/3x3/{}.txt'.format(str(file))) as f:
        lines = f.readlines() # lines is a list containing all lines of the file
        domain_id_line = lines[0]
        class_id_line = lines[1]
        user_id_line = lines[2]
        class_id_value = int(class_id_line.split()[2])
        user_id_value = int(user_id_line.split()[2])
        if not class_id_value in class_id:
            class_id[class_id_value] = []
        if not user_id_value in user_id:
            user_id[user_id_value] = []

        data = []
        for i in range(5, len(lines)):
            coords = lines[i].split(",")
            coords[3] = coords[3].strip()
            coords = [float(coord) for coord in coords]
            data.append(coords)
        
        class_id[class_id_value].extend(data)
        user_id[user_id_value].extend(data)
            
print(class_id[1])
print(class_id[2])
print(class_id[3])

print(user_id[1])
print(user_id[2])
print(user_id[3])
```

```
def distance_cost(x, y, accumulated, distances) :
    accumulated_cost = 0 
    x = class_id[1]
    y = class_id[2]
    i = len(x)-1
    j = len(y)-1


#monotonicity
    for i in range(1, len(x)) :
        i >= i - 1 
    for j in range(1, len(y)) : 
        j >= j - 1
    
    if i = i-1 and j = j : 
        accumulated_cost = accumulated[i-1, j]
    elif i = i-1 and j = j-1 :
        accumulated_cost = accumulated[i-1, j-1] 
    else : 
        accumulated_cost = accumulated[i, j-1] 
    
    distance = 0
    distances[i, j]
    
    cost = numpy.zeros((len(x), len(y)))

    for i in range(1, len(x)):
        for j in range(1, len(y)):
            cost[i, j] = min(accumulated_cost + distance[i,j] , accumulated_cost + 2*distance[i,j] , accumulated_cost + distances[i, j])
    return cost 
```
